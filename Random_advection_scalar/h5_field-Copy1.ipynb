# Structure functions

```python
shifts = np.arange(n)
dx = 2.0*np.pi/N
strfun2_long_x = np.zeros_like(shifts, dtype=float)
strfun2_long_y = np.zeros_like(shifts, dtype=float)
strfun2_long_z = np.zeros_like(shifts, dtype=float)
```

```python
shifted_array = np.zeros_like(U_phys, dtype=float)
for index in list(range(shifts.shape[0])) : 
    s=shifts[index]
    shifted_array = np.roll(scal1, s, axis=0)
    strfun2_long_x[index] = np.mean((shifted_array-scal1)**(2.0))
    shifted_array = np.roll(scal1, s, axis=1)
    strfun2_long_y[index] = np.mean((shifted_array-scal1)**(2.0))
    shifted_array = np.roll(scal1, s, axis=2)
    strfun2_long_z[index] = np.mean((shifted_array-scal1)**(2.0))
```

```python
fig, ax = plt.subplots(constrained_layout=True)

r = shifts*dx 
ax.plot(r[1::], strfun2_long_x[1::], '.-', label=r'$x$')
ax.plot(r[1::], strfun2_long_y[1::], '.-', label=r'$y$')
ax.plot(r[1::], strfun2_long_z[1::], '.-', label=r'$z$')
#ax.plot(r[1::], 3e-1*r[1::]**(2),'--')
ax.set(xscale='log', yscale='log', xlabel=r'$r/\eta$', ylabel=r'$S_2(r)$', title=r'$S_2$ of scalar 1'); ax.grid()
ax.legend(loc='best')
plt.show()
```
