# Spectrum

```python
#FFT of the velocity fields
U_spec = np.fft.rfftn(U_phys, norm='ortho')
V_spec = np.fft.rfftn(V_phys, norm='ortho')
W_spec = np.fft.rfftn(W_phys, norm='ortho')
# Wave vectors
k = np.fft.fftfreq(N_vel, d=1.0/N_vel)
kx, ky, kz = np.meshgrid(k, k, np.fft.rfftfreq(N_vel, d=1.0/N_vel), sparse=False, indexing='xy')
k_norm_array_vel = np.sqrt(kx**2 + ky**2 + kz**2)
```

```python
#FFT of the scalar fields
Scal_spec = np.empty([nb_of_scalars, N_scal, N_scal, N_scal//2+1],dtype=complex)
for sca in list(range(nb_of_scalars)) : 
    Scal_spec[sca,:,:,:] = np.fft.rfftn(scal[sca,:,:,:], norm='ortho')
    
# Wave vectors
k = np.fft.fftfreq(N_scal, d=1.0/N_scal)
kx, ky, kz = np.meshgrid(k, k, np.fft.rfftfreq(N_scal, d=1.0/N_scal), sparse=False, indexing='xy')
k_norm_array_scal = np.sqrt(kx**2 + ky**2 + kz**2)
```

## Compute Spherical spectrum

```python
dk = 0.5
k_norm_vector_vel = np.arange(1,N_vel//2-1)

velocity_spectrum = np.zeros(len(k_norm_vector_vel))

nb_of_modes_vel = np.zeros(len(k_norm_vector_vel))
k_avg_vel = np.zeros(len(k_norm_vector_vel))
```

```python
k_norm_vector_scal = np.arange(1,N_scal//2-1)
spectrum_scalar = np.zeros([nb_of_scalars,len(k_norm_vector_scal)])
dissipation_spectrum_scalar = np.zeros([nb_of_scalars,len(k_norm_vector_scal)])

nb_of_modes_scal = np.zeros(len(k_norm_vector_scal))
k_avg_scal = np.zeros(len(k_norm_vector_scal))
```

```python
for ik in list(range(len(k_norm_vector_vel))) :
    ik_norm = k_norm_vector_vel[ik]
    indices = np.nonzero((k_norm_array_vel < (ik_norm + dk)) & (k_norm_array_vel > (ik_norm - dk)))
    nb_of_modes_vel[ik] = indices[0].size
    k_avg_vel[ik] = np.sum(np.real(k_norm_array_vel[indices]))/nb_of_modes_vel[ik]
    velocity_spectrum[ik] = 0.5*np.sum(np.real(U_spec[indices]*np.conj(U_spec[indices]) + V_spec[indices]*np.conj(V_spec[indices]) + W_spec[indices]*np.conj(W_spec[indices])))
spherical_coef_vel = 4.0*np.pi*k_avg_vel**2/nb_of_modes_vel
```

```python
for ik in list(range(len(k_norm_vector_scal))) :
    ik_norm = k_norm_vector_scal[ik]
    indices = np.nonzero((k_norm_array_scal < (ik_norm + dk)) & (k_norm_array_scal > (ik_norm - dk)))
    nb_of_modes_scal[ik] = indices[0].size
    k_avg_scal[ik] = np.sum(k_norm_array_scal[indices])/nb_of_modes_scal[ik]
    for sca in list(range(nb_of_scalars)) :
        spectrum_scalar[sca,ik] = np.sum(np.real(Scal_spec[sca][indices]*np.conj(Scal_spec[sca][indices])))
        dissipation_spectrum_scalar[sca,ik] = np.sum((k_norm_array_scal[indices]**2)*np.real(Scal_spec[sca][indices]*np.conj(Scal_spec[sca][indices])))
spherical_coef_scal = 4.0*np.pi*k_avg_scal**2/nb_of_modes_scal
```

## Plot scalar spec

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5,3.5))

ax.plot(k_norm_vector_vel, velocity_spectrum/np.max(velocity_spectrum)*spherical_coef_vel, '-', color='silver', label='Velocity spec')
ax.plot(k_norm_vector_vel, 13*k_norm_vector_vel**(-1-eps),'--',color='silver', label=r'$k^{-1-\epsilon}$')
for sca in list(range(nb_of_scalars)) :
    ax.plot(k_norm_vector_scal, spectrum_scalar[sca,:]/np.max(spectrum_scalar[sca,:])*spherical_coef_scal, '-', label=r'$\kappa$ =  %2.3e' % (diffusivity[sca]))
ax.plot(k_norm_vector_scal, 30*k_norm_vector_scal**(-3+eps),'b--', label=r'$k^{-3+\epsilon}$')
ax.set(xscale='log', yscale='log', title='Scalar spectrum', xlabel='k', ylabel=r'$S_\theta(k)$'); ax.grid()
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
# pp = PdfPages('../Images/Scalar/spec_eps01_u10.pdf')
# pp.savefig(bbox_inches='tight')
# pp.close()
plt.show()
```
