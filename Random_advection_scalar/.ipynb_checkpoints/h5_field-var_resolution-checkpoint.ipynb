# Spectrum

```python
#FFT of the velocity fields
U_spec = np.fft.rfftn(U_phys, norm='ortho')
V_spec = np.fft.rfftn(V_phys, norm='ortho')
W_spec = np.fft.rfftn(W_phys, norm='ortho')
# Wave vectors
k = np.fft.fftfreq(N_vel, d=1.0/N_vel)
kx, ky, kz = np.meshgrid(k, k, np.fft.rfftfreq(N_vel, d=1.0/N_vel), sparse=False, indexing='xy')
k_norm_array_vel = np.sqrt(kx**2 + ky**2 + kz**2)
```

```python
#FFT of the scalar fields
for sca in scalars : 
    sca['spec'] = np.empty([sca['N'], sca['N'], sca['N']//2+1],dtype=complex)
    sca['spec'] = np.fft.rfftn(sca['phys_field'], norm='ortho')
    # Wave vectors
    sca['k'] = np.fft.fftfreq(sca['N'], d=1.0/sca['N'])
    kx, ky, kz = np.meshgrid(k, k, np.fft.rfftfreq(sca['N'], d=1.0/sca['N']), sparse=False, indexing='xy')
    k_norm_array_scal = np.sqrt(kx**2 + ky**2 + kz**2)
```

## Compute Spherical spectrum

```python
dk = 0.5
k_norm_vector_vel = np.arange(1,N_vel//2-1)
velocity_spectrum = np.zeros(len(k_norm_vector_vel))
nb_of_modes_vel = np.zeros(len(k_norm_vector_vel))
k_avg_vel = np.zeros(len(k_norm_vector_vel))

for ik in list(range(len(k_norm_vector_vel))) :
    ik_norm = k_norm_vector_vel[ik]
    indices = np.nonzero((k_norm_array_vel < (ik_norm + dk)) & (k_norm_array_vel > (ik_norm - dk)))
    nb_of_modes_vel[ik] = indices[0].size
    k_avg_vel[ik] = np.sum(np.real(k_norm_array_vel[indices]))/nb_of_modes_vel[ik]
    velocity_spectrum[ik] = 0.5*np.sum(np.real(U_spec[indices]*np.conj(U_spec[indices]) + V_spec[indices]*np.conj(V_spec[indices]) + W_spec[indices]*np.conj(W_spec[indices])))
spherical_coef_vel = 4.0*np.pi*k_avg_vel**2/nb_of_modes_vel
```

```python
for sca in scalars : 
    k = np.fft.fftfreq(sca['N'], d=1.0/sca['N'])
    kx, ky, kz = np.meshgrid(k, k, np.fft.rfftfreq(sca['N'], d=1.0/sca['N']), sparse=False, indexing='xy')
    k_norm_array_scal = np.sqrt(kx**2 + ky**2 + kz**2)
    k_norm = np.arange(1,sca['N']//2-1)
    sca['spherical_spec'] = np.zeros(len(k_norm))
    sca['nb_of_modes'] = np.zeros(len(k_norm))
    sca['k_avg'] = np.zeros(len(k_norm))
    for ik, ik_norm in np.ndenumerate(k_norm) :
        indices = np.nonzero((k_norm_array_scal < (ik_norm + dk)) & (k_norm_array_scal > (ik_norm - dk)))
        sca['nb_of_modes'][ik] = indices[0].size
        sca['k_avg'][ik] = np.sum(k_norm_array_scal[indices])/sca['nb_of_modes'][ik]
        sca['spherical_spec'][ik] = np.sum(np.real(sca['spec'][indices]*np.conj(sca['spec'][indices])))
    sca['spherical_coef'] = 4.0*np.pi*sca['k_avg']**2/sca['nb_of_modes']   
```

## Plot scalar spec

# Structure functions

```python
# function calculates structure function of a field of order p in 3 directions (x,y,z)
def structure_functions(field, p) :
    N = field.shape[0]
    shifts = np.arange(N)
    shifted_array = np.zeros_like(field, dtype=float)
    strfun2_long_x = np.zeros_like(shifts, dtype=float)
    strfun2_long_y = np.zeros_like(shifts, dtype=float)
    strfun2_long_z = np.zeros_like(shifts, dtype=float)
    for index in list(range(N)) : 
        s=shifts[index]
        shifted_array = np.roll(field, s, axis=0)
        strfun2_long_x[index] = np.mean((shifted_array-field)**p)
        shifted_array = np.roll(field, s, axis=1)
        strfun2_long_y[index] = np.mean((shifted_array-field)**p)
        shifted_array = np.roll(field, s, axis=2)
        strfun2_long_z[index] = np.mean((shifted_array-field)**p)
    return strfun2_long_x, strfun2_long_y, strfun2_long_z
```

## Velocity

```python
[U_strfun2_along_x,U_strfun2_along_y, U_strfun2_along_z] = structure_functions(U_phys,2.0)
[V_strfun2_along_x,V_strfun2_along_y, V_strfun2_along_z] = structure_functions(V_phys,2.0)
[W_strfun2_along_x,W_strfun2_along_y, W_strfun2_along_z] = structure_functions(W_phys,2.0)
r_vel = np.arange(N_vel)*2.0*np.pi/N_vel
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, constrained_layout=True,figsize=(15,4))
compensation = r_vel[1::]**(0)
ax[0].plot(r_vel[1::], U_strfun2_along_x[1::]*compensation, '.-', label=r'$x$')
ax[0].plot(r_vel[1::], U_strfun2_along_y[1::]*compensation, '.-', label=r'$y$')
ax[0].plot(r_vel[1::], U_strfun2_along_z[1::]*compensation, '.-', label=r'$z$')
ax[0].plot(r_vel, 10*r_vel**(eps), '-')
ax[0].set(xscale='log', yscale='log', xlabel=r'$r$', ylabel=r'$S_2(r)/r^\epsilon$', title=r'$S_2$ of velocity U'); ax[0].grid()

ax[1].plot(r_vel[1::], V_strfun2_along_x[1::]*compensation, '.-', label=r'$x$')
ax[1].plot(r_vel[1::], V_strfun2_along_y[1::]*compensation, '.-', label=r'$y$')
ax[1].plot(r_vel[1::], V_strfun2_along_z[1::]*compensation, '.-', label=r'$z$')
ax[1].plot(r_vel, 10*r_vel**(eps), '-')
ax[1].set(xscale='log', yscale='log', xlabel=r'$r$', ylabel=r'$S_2(r)/r^\epsilon$', title=r'$S_2$ of velocity V'); ax[1].grid()

ax[2].plot(r_vel[1::], W_strfun2_along_x[1::]*compensation, '.-', label=r'$x$')
ax[2].plot(r_vel[1::], W_strfun2_along_y[1::]*compensation, '.-', label=r'$y$')
ax[2].plot(r_vel[1::], W_strfun2_along_z[1::]*compensation, '.-', label=r'$z$')
ax[2].plot(r_vel, 10*r_vel**(eps), '-')
ax[2].set(xscale='log', yscale='log', xlabel=r'$r$', ylabel=r'$S_2(r)/r^\epsilon$', title=r'$S_2$ of velocity W'); ax[2].grid()
ax[2].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
