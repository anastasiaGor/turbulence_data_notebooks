```python
folder = "/home/gorbunov8a/Documents/MATLAB/data/spectres/spectra/"
filetype = "_spectrum_kstar_avg.table"

for i in list(range(len(simu))) :
    file = folder + "rl" + str(simu[i]['Rl']) + "_N" + str(simu[i]['N']) + filetype
    file_data = []
    for line in open(file) :
        temp_array_line = []
        if line[0] == '#' :          #skip commented lines
            continue
        else :
            temp_array_line=list(map(float, line.split( )))
            file_data.append(temp_array_line)
    simu[i].update({'lambda': file_data[0][0], 'eta': file_data[0][1], 'epsilon' : file_data[0][2], \
                    'table' : np.array(file_data[1:]), \
                    'legend' : r"$R_{\lambda}=$" + str(simu[i]['Rl']) + r"$, N = $" + str(simu[i]['N']), \
                    'legend_short' : r"$R_{\lambda}=$" + str(simu[i]['Rl'])})
del temp_array_line, file_data
```

# Fitting of derivatives

## 3rd derivative

```python
n=4 #number of points to eliminate in the end
deriv3_fit=np.zeros((len(simu),2),dtype=float)
for i in good_resolution_simu :
    wavenumbers = simu[i]['table'][:,0]
    energy = simu[i]['table'][:,1]
    deriv1 = np.gradient(np.log(energy),np.log(wavenumbers*simu[i]['eta']), edge_order=1)
    deriv2 = np.gradient(deriv1,np.log(wavenumbers*simu[i]['eta']), edge_order=1)
    deriv3 = np.gradient(np.log(np.absolute(deriv2)),np.log(wavenumbers*simu[i]['eta']), edge_order=1)
    deriv3 = deriv3[:-n]
    # find the corresponding indexes in the wavenumber vector
    a = np.searchsorted(wavenumbers[:-n]*simu[i]['eta'], keta_min)
    b = np.searchsorted(wavenumbers[:-n]*simu[i]['eta'], keta_max)
    x = wavenumbers[a:b]*simu[i]['eta']
    y = deriv3[a:b]
    deriv3_fit[i,0] = np.mean(y)
    deriv3_fit[i,1] = np.sqrt(np.sum((y-deriv3_fit[i,0])**2))
```

## 2nd derivative

```python
n=3 #number of points to eliminate in the end
deriv2_fit=np.zeros((len(simu),2),dtype=float)
for i in good_resolution_simu :
    wavenumbers = simu[i]['table'][:,0]
    energy = simu[i]['table'][:,1]
    deriv1 = np.gradient(np.log(energy),np.log(wavenumbers*simu[i]['eta']), edge_order=1)
    deriv2 = np.gradient(deriv1,np.log(wavenumbers*simu[i]['eta']), edge_order=1)
    deriv2 = deriv2[:-n]
    # find the corresponding indexes in the wavenumber vector
    a = np.searchsorted(wavenumbers[:-n]*simu[i]['eta'], keta_min)
    b = np.searchsorted(wavenumbers[:-n]*simu[i]['eta'], keta_max)
    x = np.log(wavenumbers[a:b]*simu[i]['eta'])
    y = np.log(-1.0*(deriv2[a:b]))
    deriv2_fit[i] = np.polyfit(x,y,1)
```

## 1st derivative

```python
def deriv1_function(x, a, b, mu):   
    #a=2.0/3.0
    return -b - mu*a*np.exp(a*x)
nb_of_params = 3

n=2 #number of points to eliminate in the end
deriv1_fit=np.zeros((len(simu),nb_of_params),dtype=float)
cov_matrix1=np.zeros((len(simu),nb_of_params,nb_of_params),dtype=float)
for i in good_resolution_simu :
    wavenumbers = simu[i]['table'][:,0]
    energy = simu[i]['table'][:,1]
    deriv1 = np.gradient(np.log(energy),np.log(wavenumbers*simu[i]['eta']), edge_order=1)
    deriv1 = deriv1[:-n]
    # find the corresponding indexes in the wavenumber vector
    a = np.searchsorted(wavenumbers[:-n]*simu[i]['eta'], keta_min)
    b = np.searchsorted(wavenumbers[:-n]*simu[i]['eta'], keta_max)
    x = np.log(wavenumbers[a:b]*simu[i]['eta'])
    y = deriv1[a:b]
    deriv1_fit[i,:], cov_matrix1[i, :, :] = curve_fit(deriv1_function,x,y)
```

```python
def stretched_exponent(x, a, b, mu, c):   
    #a=2.0/3.0
    return c - b*x - mu*np.exp(a*x)
nb_of_params = 4

n=1 #number of points to eliminate in the end
deriv0_fit=np.zeros((len(simu),nb_of_params),dtype=float)
cov_matrix0=np.zeros((len(simu),nb_of_params,nb_of_params),dtype=float)
for i in good_resolution_simu :
    wavenumbers = simu[i]['table'][:,0]
    energy = simu[i]['table'][:,1]
    energy = energy[:-n]
    # find the corresponding indexes in the wavenumber vector
    a = np.searchsorted(wavenumbers[:-n]*simu[i]['eta'], keta_min)
    b = np.searchsorted(wavenumbers[:-n]*simu[i]['eta'], keta_max)
    x = np.log(wavenumbers[a:b]*simu[i]['eta'])
    y = np.log(energy[a:b])
    deriv0_fit[i,:], cov_matrix0[i, :, :] = curve_fit(stretched_exponent,x,y)
```
