```python
#%matplotlib notebook
import numpy as np
import scipy.fftpack
import matplotlib.pyplot as plt
```

```python
N = 1001
delta = 2/(N-1)
x = np.linspace(-(N-1)//2, (N-1)//2, N)*delta
nu_forced = 5.0
y = np.exp(nu_forced *1j *2.0*np.pi* x) 
plt.plot(x,np.real(y), '.-')
plt.plot(x,np.real(np.conj(np.flip(y))), '.-')
plt.plot(x,np.imag(y), '.-')
plt.plot(x,np.imag(np.conj(np.flip(y))), '.-')
```

```python
plt.plot(np.real(np.conj(np.flip(y)) - y))   # check Hermitian symmetry
plt.plot(np.imag(np.conj(np.flip(y)) - y)) 
```

```python
nu = np.fft.fftfreq(N, delta)
nu = np.fft.fftshift(nu)
yf = np.fft.fft(y)
a=-(N-1)//2
shift_vector = np.exp(-1.0j * 2*np.pi*a*(np.arange(N))/(N))
#print(shift_vector)
yf_shifted = np.multiply(yf, shift_vector)
#yf_shifted = yf

#plt.plot(nu, 1.0/N*np.abs(np.fft.fftshift(yf_shifted)))
plt.plot(nu, 1.0/N*np.imag(np.fft.fftshift(yf_shifted)))
plt.xlim(-20, 20) 
plt.show()
```
