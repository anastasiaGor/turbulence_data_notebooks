# Import

```python
import numpy as np
import matplotlib.pyplot as plt
import time as time
from scipy import sparse
import scipy
from scipy.sparse.linalg import spsolve
```

# Méthodes iteratives

```python
def relax(ksolve, ksch, N, Te):  
    niter=0  # nombre d'itération des méthodes itératives
    nitermax = 20000  # nombre maxi d'iteration de jacobi
    restab = np.zeros(nitermax)
    errtab = np.zeros(nitermax)
    pi = np.pi
    #
    # Generation du maillage selon les axes x et y
    #
    x = np.linspace(0, 1, N)
    y = np.linspace(0, 1, N)
#
#    solution approchee  T
    T = np.zeros([N, N])

    nitermax = 20000  #  nombre maxi d'iteration de jacobi
#
    om = 2. * (1 - np.pi/N) # parametre de surrelaxation omega
#
#  Application des conditions aux limites (valeurs imposees en debut de calcul)
    T[0, :] = 0.
    T[:, 0] = 0.
    for i in range(0, N):
         T[N-1, i] = np.sin(pi*y[i])

    for i in range(0, N):
        T[i, N-1] = np.sin(pi*x[i])

    res = 1 #résidu
    while res > 1.0E-8 and niter < nitermax:  # on continu les iterations tant que le residu est trop grand
        #  stockage du du niveau N+1 au niveau N
        U = T.copy()
        if ksolve == 1:  # méthode de jacobi	    
            if ksch == 1:
                #  methode de jacobi a 5 Pts
                for i in range(1, N-1):
                    for j in range(1, N-1):
                        T[i, j] = 0.25*(U[i+1, j] + U[i-1, j] + U[i, j+1] + U[i, j-1])

            if ksch == 2:
                #  methode de jacobi a 9 pts
                for i in range(1, N - 1):
                    for j in range(1, N - 1):
                        T[i, j]= 0.2*(U[i+1,j]+U[i-1,j]+U[i,j+1]+U[i,j-1])+ 0.05*(U[i+1,j+1]+U[i-1,j-1]+U[i+1,j-1]+U[i-1,j+1])

        if ksolve == 2:  # méthode de Gauss Seidel
            #print ("Gauss-Seidel non implémenté")
            #return 0,[1],[0],T
            if ksch == 1:
                #  methode GS a 5 Pts
                for i in range(1, N-1):
                    for j in range(1, N-1):
                        T[i, j] = 0.25*(U[i+1, j] + T[i-1, j] + U[i, j+1] + T[i, j-1])               
            if ksch == 2:
                for i in range(1, N-1):
                    for j in range(1, N-1):
                        T[i, j] = 0.2*(U[i + 1, j] + T[i - 1, j] + U[i, j + 1] + T[i, j - 1]) + 0.05*(U[i + 1, j+1] + T[i - 1, j+1] + U[i+1, j - 1] + T[i-1, j - 1])
              
        if ksolve == 3:  # méthode SOR
            #print ("SOR non implémenté")
            #return 0,[1],[0],T
            if ksch == 1:
                #  methode SOR a 5 Pts                
                for i in range(1, N-1):
                    for j in range(1, N-1):
                        T[i, j] = (1 - om)*U[i,j] + om*0.25*(U[i+1, j] + T[i-1, j] + U[i, j+1] + T[i, j-1])  
            if ksch == 2:
                for i in range(1, N-1):
                    for j in range(1, N-1):
                        T[i, j] = (1 - om)*U[i,j] + om*(0.2*(U[i + 1, j] + T[i - 1, j] + U[i, j + 1] + T[i, j - 1]) + 0.05*(U[i + 1, j+1] + T[i - 1, j+1] + U[i+1, j - 1] + T[i-1, j - 1]))
                

        niter = niter+1
       
        # Calcul du residu en norme L2
        res = np.linalg.norm(T-U)/N
        restab[niter-1] = np.log10(res)

        # calcul de l'erreur en norme L2
        err = np.linalg.norm(T-Te)/N
        errtab[niter-1] = np.log10(err)
        
        

    return (niter, errtab, restab, T)
```

# Méthodes directes

```python
####  fonction construisant la matrice et le second membre ####
def systeme(ksolve,ksch,N):
    if ksolve <= 4: #matrice ordinaire
        A = np.zeros([N * N, N * N])
    else: #ksolve=5: matrice creuse, le format "lil" est le plus rapide en remplissage
        A = sparse.lil_matrix((N * N, N * N))  # ne marche pas avec []

    #
    # Generation du maillage selon les axes x et y
    #
    x = np.linspace(0, 1, N)
    y = np.linspace(0, 1, N)

    if ksch == 1:    # Remplissage dans le la matrice A dans le cas du schéma à  5 points
        for i in range(0, N):
            A[i, i] = 1

        for i in range(1, N-1):
            ii = i*N
            A[ii, ii] = 1
            for j in range(1, N-1):
                ii = i*N + j
                A[ii, ii] = 1
                A[ii, ii - 1] = -1. / 4.
                A[ii, ii + 1] = -1. / 4.
                A[ii, ii - N] = -1. / 4.
                A[ii, ii + N] = -1. / 4.

            ii = i*N + N - 1
            A[ii, ii] = 1

        for i in range(N*(N-1)-1, N * N):
            A[i, i] = 1

    else:
        # Remplissage de la matrice A dans le cas du schéma à  9 points 
        for i in range(0, N):
            A[i, i] = 1

        for i in range(1, N-1):
            ii = i*N
            A[ii, ii] = 1
            for j in range(1, N-1):
                ii = i* N + j
                A[ii, ii] = 1
                A[ii, ii - 1] = -1. / 5.
                A[ii, ii + 1] = -1. / 5.
                A[ii, ii - N] = -1. / 5.
                A[ii, ii + N] = -1. / 5.
                A[ii, ii - 1 - N] = -1. / 20.
                A[ii, ii + 1 - N] = -1. / 20.
                A[ii, ii - 1 + N] = -1. / 20.
                A[ii, ii + 1 + N] = -1. / 20.

            ii = i*N + N - 1
            A[ii, ii] = 1

        for i in range(N*(N-1)-1, N * N):
            A[i, i] = 1

    # Remplissage du vecteur b
    b = np.zeros(N * N)

    for i in range(N):
        ii = i*N + N -1
        b[ii] = np.sin(pi * x[i])

    for i in range(N):
        ii = (N - 1) * N + i
        b[ii] = np.sin(pi * y[i])
        
    return A,b
    
####  fonction résolvant le systà¨me  ####
def direct(ksolve, ksch , N, Te):  #
    A,b = systeme(ksolve,ksch,N)    

    # Résolution du systà¨me par méthode LU
    #matrice ordinaire:
    if ksolve == 4:
        resu = np.linalg.solve(A, b)
    #matrice creuse:
    else:
        resu = scipy.sparse.linalg.spsolve (scipy.sparse.csr_matrix(A), b)

    # Remise en forme de la solution
    T = np.zeros([N, N])
    ii = 0
    for i in range(0,N):
        for j in range(0,N):
            T[i, j] = resu[ii]
            ii = ii + 1

    err = np.linalg.norm(T-Te)/N
    return (err,T)
```

# Parameters de résolution

Choix du schema : 1 - 5pts, 2 - 9pts

```python
ksch = 2
```

Méthodes de résolution: 
=1 -> Jacobi
=2 -> Gauss-Seidel
=3 -> SOR
=4 -> méthode de pivot (factorisation LU générale)
=5 -> factorisation LU pour matrice creuse 

valeurs prises par la variable ksolve ci-après, enlever les méthodes itératives pour les grands N

solveurs = [1,2,3,4,5]

```python
solveurs=[4,5]
```

Valeurs de N à tester :

```python
tailles=[20,40,60,80,160]
#tailles=[20]
```

```python
pi = np.pi
```

# Résolution

```python
print ("ksch | ksolve | N | log10 du pas | log10 de l'erreur | niter | temps cpu (s)")
for ksolve in solveurs:
    niter=0
    for N in tailles:
        # Generation du maillage selon les axes x et y
        #
        h = 1. / (N - 1)  # Pas d'espace dx, dy
        x = np.linspace(0, 1, N)
        y = np.linspace(0, 1, N)
        #
        # La solution exacte
        #
        Te = np.zeros([N, N])
        for i in range(0, N):
            for j in range(0, N):
                Te[i, j] = (1 / np.sinh(np.pi)) * (
                    np.sinh(np.pi * x[i]) * np.sin(np.pi * y[j]) + np.sin(np.pi * x[i]) * np.sinh(np.pi * y[j]))
        
        debtime = time.time()
        if ksolve < 4:  # méthodes itératives
            niter, errtab, restab, T = relax(ksolve, ksch, N, Te)
            err = errtab[niter-1]
            #print ksch,'\t',ksolve,'\t',N,'\t',np.log10(h),'\t', errtab[niter-1],'\t',niter,'\t',time.time()-debtime
        elif ksolve<6:
            err,T =direct(ksolve, ksch, N, Te)
            niter=1
            err = np.log10(err)
            #print ksch,'\t',ksolve,'\t',N,'\t',np.log10(h),'\t', np.log10(err),'\t',niter,'\t',time.time()-debtime
        else: # dernière question uniquement
            import jacobiOptimise as jo
            if ksolve<=7:
                niter, errtab, restab, T = jo.relaxMatrice(ksolve, ksch, N, Te)
            if ksolve==8:
                niter, errtab, restab, T = jo.relaxTableau(ksolve, ksch, N, Te)
            err = errtab[niter-1]
        
        print (ksch,'\t',ksolve,'\t',N,'\t',np.log10(h),'\t', err,'\t',niter,'\t',time.time()-debtime)
```
