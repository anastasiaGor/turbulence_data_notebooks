```python
import numpy as np

import matplotlib.pylab as plt
```

```python
# Affectation des constantes

# Parametres physiques

D = 30.  # diffusivite de la paroi(en m2 / s)
L = 1.  # epaisseur de la plaque(en m)
cg = 0.1  # concentrationface gauche(en K)
cd = 0.  # concentration face droite(enK)
```

```python
# Parametres numeriques

pasespace=0.01   # pas d 'espace  (en m)
```

```python
# Maillage numerote de i=0 (x=0) a i=N-1 (x=L) -> N noeuds interieurs + exterieur

npoint=int(1+L/pasespace)
lamda=0.49  # parametre sans dimension: lambda = alpha * dt / (dx * dx)  - > dt
tempsfinal=0.001
```

```python
# Parametres deduits

pastemps = lamda * pasespace * pasespace / D  # pas de temps (en s)

itermax = int(tempsfinal /pastemps)  # nombre d'iterations (en temps)
```

```python
# Maillage numerote de i=0 (x=0) a i=N-1 (x=L) -> N noeuds interieurs + exterieur

# Stockage des champs

cold =  np.zeros(npoint)  # vecteur concentration dans le tube
cold[0] = cg
cold[npoint-1] = cd
c = np.zeros(npoint)

evo = np.zeros(itermax + 1)  # vecteur profil de concentration fonction du temps taille nombre iteration +1 (valeur initiale)
```

```python
# Iterations----------------------------

for i in range(0, itermax):  # itermax iteration avec i variant de 0 a itermax -1
    for j in range(1, npoint-1):  # mise a jour des noeuds internes
        c[j] = lamda *cold[j-1] +lamda *cold[j+1] +(1-2*lamda)*cold[j]
    c[0] = cg  # inutile mais plus propre
    c[npoint - 1] = cd  # idem
    evo[i + 1] = c[int((npoint + 1) / 2)]  # valable qu'avec N impair  pour avoir un noeus au milieu
    cold = np.copy(c)
```

```python
# Trace champ de concentration - -----------

x = np.linspace(0, L, npoint)

# Trace profil de concentration - ----------
t = np.linspace(0, tempsfinal, itermax + 1)

plt.plot(x, c, 'r--')
plt.xlabel("coordonnee x")
plt.ylabel("concentration finale gaz")
fig = plt.figure()
plt.plot(t, evo)
plt.xlabel("temps")
plt.ylabel("concentration finale gaz")
```

```python
# Definition de la matrice - -------------
DF = np.zeros((npoint, npoint))

DF[0, 0] = 1
DF[npoint - 1, npoint -1] = 1
for i in range(1, npoint - 1):
    DF[i, i] = 1 - 2 * lamda
    DF[i, i + 1] = lamda
    DF[i, i - 1] = lamda
# --------------------------------------
```

```python
c1 = np.zeros(npoint)  # vecteur concentration dans la plaque
c1[0] = cg
c1[npoint-1] = cd
```

```python
for i in range(0, itermax):  # itermax iteration avec i variant de 0 a itermax -1
    c1 = DF.dot(c1)
    evo[i + 1] = c1[int((npoint + 1) / 2)]  # valable qu'avec N impair  pour avoir un noeus au milieu
```

```python
fig = plt.figure()
plt.plot(x, c1, 'r--')
plt.xlabel("coordonnee x")
plt.ylabel("concentration finale gaz")
fig = plt.figure()
plt.plot(t, evo)
plt.xlabel("temps")
plt.ylabel("concentration finale gaz")

plt.show()
```
