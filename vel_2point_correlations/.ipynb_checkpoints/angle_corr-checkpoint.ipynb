```python
#reading routine
def read_table(folder, file) :
    file = folder + file
    file_list = []
    print(file)
    for line in open(file) :
        temp_array_line = []
        try :
            if line[0] == '#' :          #skip commented lines
                continue
            else :
                temp_array_line=list(map(float, line.split( )))
                file_list.append(temp_array_line)
        except :
            print('error')
    k = np.array(file_list[0])
    nb_of_modes = np.array(file_list[1])
    table = np.array(file_list[2:])
    t = table[:,0]
    corr = table[:,1:]
    return [k, nb_of_modes,t,corr]
```

```python
#reading routine
def read_vector(folder, file) :
    file = folder + file
    file_list = []
    print(file)
    for line in open(file) :
        temp_array_line = []
        try :
            if line[0] == '#' :          #skip commented lines
                continue
            else :
                temp_array_line=list(map(float, line.split( )))
                file_list.append(temp_array_line)
        except :
            print('error')
    table = np.array(file_list)
    t = table[:,0]
    vector = table[:,1]
    return [t, vector]
```

```python
Rl = 90; N = 256
[k, nb_of_modes, t, transfer] = read_table("/home/gorbunov8a/data/correlations/rl90_N256/angle/", "total_transfer_corr_lin_Cavg_wi=025.table")
[t_pos, angle_pos] = read_vector("/home/gorbunov8a/data/correlations/rl90_N256/angle/", "angle_velocity_lamb_positive_time_wi=025.table")
[t_neg, angle_neg] = read_vector("/home/gorbunov8a/data/correlations/rl90_N256/angle/", "angle_velocity_lamb_negative_time_wi=025.table")
```

```python
angle = np.concatenate((np.flipud(angle_neg), angle_pos))
t_angle = np.concatenate((np.flipud(t_neg), t_pos))
```

```python
total_transfer = np.sum(transfer, axis=1)
```

```python
fig, (ax) = plt.subplots(nrows=1, ncols=1,figsize=(5,4), constrained_layout=True)
ax.plot(t_angle, angle/np.max(np.abs(angle)), label='angle')
ax.plot(t, total_transfer/np.max(total_transfer), label='transfer')
ax.grid()
ax.legend()
plt.show()
```
