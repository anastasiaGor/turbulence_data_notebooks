```python
#reading routine
def read_table(folder, file) :
    file = folder + file
    file_list = []
    print(file)
    for line in open(file) :
        temp_array_line = []
        try :
            if line[0] == '#' :          #skip commented lines
                continue
            else :
                temp_array_line=list(map(float, line.split( )))
                file_list.append(temp_array_line)
        except :
            print('error')
    k = np.array(file_list[0])
    nb_of_modes = np.array(file_list[1])
    table = np.array(file_list[2:])
    t = table[:,0]
    corr = table[:,1:]
    return [k, nb_of_modes,t,corr]
```

Spatial spectrum :

```python
fig, ax = plt.subplots(constrained_layout=True, nrows=1, ncols=1, figsize=(4,3.5))
ax.plot(k, real_corr[0,:]*4.0*np.pi*k**2, '.-')
ax.plot(k, 10e-4*k**(-5.0/3.0))
ax.set(title=r'Spatial spectrum',xlabel=r'$k$', xscale='log', yscale='log', ylabel=r'$4 \pi k^2 C_2(0, k)$'); ax.grid()

# pp = PdfPages('Images/Scalar/delta_tk.pdf')
# pp.savefig(bbox_inches='tight')
# pp.close()
plt.show()
```

```python
delta = t[1]-t[0]
t_length = len(t)
omega = np.fft.fftfreq(t_length, delta)*2.0*np.pi
spectrum_of_correlation = np.ndarray(shape=(len(t), len(k)), dtype=complex)
for ik in list(range(len(k))) :
    yf = np.fft.fft(real_corr[:,ik])
    spectrum_of_correlation[:, ik] = 1.0/len(t) * yf
```

```python
#plot \hat{C}_2(w, k) = FFT [C_2(t, k)]
fig, ax = plt.subplots(constrained_layout=True, figsize=(5.5,3.5))
for ik in indices_of_k_to_plot :
    ax.plot(omega[0:t_length//2], np.absolute(spectrum_of_correlation[0:t_length//2, ik]), '.-', label='k=%2.f' % (k[ik]))
    #plt.plot(nu, 1.0/N * np.imag(yplot[:, ik]), label='k=%2.f' % k[ik])
#plt.xlim(-15, 15) 
#plt.plot(nu[1:N//2]*horizontal_scale, 10e-2*nu[1:N//2]**(-1), 'k-', label=r'$\nu^{-1}$')
ax.set(xscale='log', yscale='log', xlabel=r'$\omega^\prime$', ylabel=r'$\hat{C}_2 (\omega, k)$', title='FFT of autocorrelation for different wavenumbers')
ax.grid()
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#plot SCALED \hat{C}_2(w, k) = FFT [C_2(t, k)]
fig, ax = plt.subplots(constrained_layout=True, figsize=(5.5,3.5))
for ik in indices_of_k_to_plot :
    vertical_scale = 1.0/np.absolute(spectrum_of_correlation[0, ik])
    vertical_scale = 1.0
    horizontal_scale = 1.0/k[ik]
    ax.plot(omega[1:t_length//2]*horizontal_scale, np.absolute(spectrum_of_correlation[1:t_length//2, ik])*vertical_scale, '.-', label='k=%2.f' % (k[ik]))
ax.set(xscale='linear', yscale='log', xlabel=r'$\omega^\prime / k^\prime$', ylabel=r'$\hat{C}_2 (\omega, k) \ / \ \hat{C}_2 (\omega=0, k)$', title='FFT of autocorrelation for different wavenumbers \n SCALED')
ax.grid()
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
fig, ax = plt.subplots(constrained_layout=True)
for ik in indices_of_k_to_plot :
    spectrum = np.absolute(spectrum_of_correlation[:, ik])
    norm_coef = spectrum[0]
    horizontal_scale = 1.0/k[ik]
    #horizontal_scale = 1.0
    ax.plot(omega[1:t_length//2]*horizontal_scale, -np.log(spectrum[1:t_length//2]/norm_coef), '.-', label='k=%2.f' % (k[ik]))
    #plt.plot(nu, 1.0/N * np.imag(yplot[:, ik]), label='k=%2.f' % k[ik])
#plt.xlim(-15, 15) 
plt.plot(omega[1:t_length//2]*horizontal_scale, 10e3*np.power(omega[1:t_length//2]*horizontal_scale, 2.0), 'k--', label=r'$(\omega/k)^2$')
plt.plot(omega[1:t_length//2]*horizontal_scale, 2e1*np.power(omega[1:t_length//2]*horizontal_scale, 0.5), 'r--', label=r'$(\omega/k)^{0.5}$')
ax.set(xscale='log', yscale='log', xlabel=r'$\omega^\prime / k^\prime$',title='Log of FFT of autocorrelation')
ax.grid()
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
fig, ax = plt.subplots(constrained_layout=True)
total_energy_spectrum = np.zeros(shape=spectrum_of_correlation.shape[0], dtype=float)
range_of_k = list(range(len(k)))
for ik in range_of_k : 
    spectrum = np.absolute(spectrum_of_correlation[:, ik])
    coefficient = 4.0*np.pi*k[ik]*k[ik]
    #coefficient = 1.0
    total_energy_spectrum = total_energy_spectrum + spectrum*coefficient
    #total_energy_spectrum = total_energy_spectrum + spectrum
total_energy_spectrum = total_energy_spectrum/len(range_of_k)
ax.plot(omega[1:t_length//2], total_energy_spectrum[1:t_length//2], '.-', label='Spectrum')
ax.plot(omega[1:t_length//2], 2.0e-5*omega[1:t_length//2]**(-5.0/3.0),label=r'$\omega^{-5/3}$')
ax.plot(omega[1:t_length//2], 10e-5*omega[1:t_length//2]**(-1.0), label=r'$\omega^{-1}$')

ax.set(xscale='log', yscale='log', xlabel=r'$\omega^\prime$', title='FFT of total energy'); ax.grid()
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    

plt.show()
```
